```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from scipy import interpolate

from simla_variables import SimlaVar
from simladb import query, DB_bcd
from simla_utils import bcd_spectrum
```

```python
simlapath = SimlaVar().simlapath
sl_ramptimes = SimlaVar().sl_ramptimes
ll_ramptimes = SimlaVar().ll_ramptimes
zodipath = SimlaVar().zodipath
irspath = SimlaVar().irspath
plotpath = simlapath+'validations/plots/'
```

```python
eval_lams = [10.95, None, 28.75] # SL, LL, micron
zodi_lams = [np.load(zodipath+'sl_wavelengths.npy'), None, np.load(zodipath+'ll_wavelengths.npy')]
```

```python
q = query(DB_bcd.select(DB_bcd.FILE_NAME, DB_bcd.CHNLNUM, DB_bcd.RAMPTIME)\
         .where(((DB_bcd.CHNLNUM==0)|(DB_bcd.CHNLNUM==2))&(DB_bcd.OBJECT=='NCVZ-dark-spot')))
fnames, chnlnums, ramptimes = \
    irspath+q['FILE_NAME'].to_numpy(), q['CHNLNUM'].to_numpy(), q['RAMPTIME'].to_numpy()
```

```python
extractor = bcd_spectrum()
```

# Agreement Between Model and Time-dependent CVZ Intensity

```python
for chnlnum in [0, 2]:
    
    if chnlnum == 0: ramps = sl_ramptimes
    elif chnlnum == 2: ramps = ll_ramptimes
    for ramp in ramps:

        cvzs = fnames[np.where((chnlnums==chnlnum)&(np.abs(ramptimes-ramp)<0.01))]
        x, y, z = [], [], []

        for cvz in cvzs:

            imdat = fits.getdata(cvz)
            imhead = fits.getheader(cvz)
        
            mjd = imhead['MJD_OBS']
            aor = imhead['AORKEY']
            campaign = imhead['CAMPAIGN']

            l, f = extractor.fullslit_bcd_spectrum(imdat, imhead, 1) 

            modname = ['sl', None, 'll'][chnlnum]
            eval_lam = eval_lams[chnlnum]
            zodilam = zodi_lams[chnlnum]
            
            zodi_spec = np.load(zodipath+str(aor)+'_'+modname+'_zodionly.npy')
            camp_cvz = np.load(simlapath+'storage/cvz_campaign_spectra/'+'CVZ_'+modname+'_'+str(ramp)+'_'+campaign+'.npy')
        
            darkspec_model = zodi_spec - camp_cvz

            x.append(mjd)
            y.append(np.nanmedian(f[np.where((l>=eval_lam-0.5)&(l<=eval_lam+0.5))]))
            z.append(np.nanmedian(darkspec_model[np.where((zodilam>=eval_lam-0.5)&(zodilam<=eval_lam+0.5))]))

        sort_dat = sorted([[x[i], y[i], z[i]] for i in range(len(x)) if -5 < y[i] < 5])
        px = [i[0] for i in sort_dat]
        py = [i[1] for i in sort_dat]
        pz = [i[2] for i in sort_dat]

        plt.rcParams["figure.figsize"] = (12, 5)
        plt.hexbin(x=px, y=py, cmap='gray_r', bins='log', mincnt=1, gridsize=70)
        plt.plot(px, pz, color='red')
        plt.title(['SL', None, 'LL'][chnlnum]+' RAMPTIME:'+str(ramp)+'s')
        plt.ylabel(str(eval_lam)+'um Spectral Intensity (MJy/sr)')
        plt.xlabel('MJD')
        plt.colorbar()
        plt.savefig(plotpath+'/cvz/cvzplot_'+['SL', None, 'LL'][chnlnum]+'_'+str(ramp)+'.pdf', format='pdf')
        plt.show()
```

```python
for chnlnum in [0, 2]:
    
    if chnlnum == 0: ramps = sl_ramptimes
    elif chnlnum == 2: ramps = ll_ramptimes
    for ramp in ramps:

        cvzs = fnames[np.where((chnlnums==chnlnum)&(np.abs(ramptimes-ramp)<0.01))]
        x, y = [], []

        for cvz in cvzs:

            imdat = fits.getdata(cvz)
            imhead = fits.getheader(cvz)
        
            mjd = imhead['MJD_OBS']
            aor = imhead['AORKEY']
            campaign = imhead['CAMPAIGN']

            l1, f1 = extractor.fullslit_bcd_spectrum(imdat, imhead, 1) 
            l2, f2 = extractor.fullslit_bcd_spectrum(imdat, imhead, 2)
            full_l, full_f = l2.tolist()+l1.tolist(), f2.tolist()+f1.tolist()
            full_extract = sorted([[full_l[i], full_f[i]] for i in range(len(full_l))])
            l, f = np.asarray([i[0] for i in full_extract]), np.asarray([i[1] for i in full_extract])

            modname = ['sl', None, 'll'][chnlnum]
            eval_lam = eval_lams[chnlnum]
            zodilam = zodi_lams[chnlnum]
            
            zodi_spec = np.load(zodipath+str(aor)+'_'+modname+'_zodionly.npy')
            camp_cvz = np.load(simlapath+'storage/cvz_campaign_spectra/'+'CVZ_'+modname+'_'+str(ramp)+'_'+campaign+'.npy')
            darkspec_model = zodi_spec - camp_cvz
            sort_zodi = sorted([[zodilam.tolist()[i], darkspec_model.tolist()[i]] for i in range(len(zodilam))])
            zl, zf = np.asarray([i[0] for i in sort_zodi]), np.asarray([i[1] for i in sort_zodi])
            interp_zf = interpolate.interp1d(zl, zf, bounds_error=False, fill_value=0)(l)

            x.extend(l)
            y.extend(f-interp_zf)

        x, y = np.asarray(x), np.asarray(y)
        m = np.where(np.abs(y)<10)
        px, py = x[m], y[m]

        plt.rcParams["figure.figsize"] = (12, 5)
        plt.hexbin(x=px, y=py, cmap='gray_r', bins='log', mincnt=1, gridsize=70)
        plt.title(['SL', None, 'LL'][chnlnum]+' RAMPTIME:'+str(ramp)+'s')
        plt.ylabel('CVZ Obs - Corrected Zodi Model (MJy/sr)')
        plt.xlabel('micron')
        plt.colorbar()
        plt.savefig(plotpath+'/cvz/specplot_'+['SL', None, 'LL'][chnlnum]+'_'+str(ramp)+'.pdf', format='pdf')
        plt.show()
```
